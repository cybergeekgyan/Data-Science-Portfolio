**Web Scrapping**

```python
## Importing Libraries

import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://en.wikipedia.org/wiki/List_of_Academy_Award-winning_films"
```

```python
resp = requests.get(url)
```

```python
## Response code 200 -> OK
## Any status from 200 is an error

resp
```

```python
resp.status_code
```

```python
soup = BeautifulSoup(resp.content, 'lxml')
```

```python
soup
```

```python
soup.prettify
```

```python
arr = []
for i in soup.findAll('td'):
  arr.append(str(i))
```

```python
arr
```

```python
arr[0]
```

### + -> First or more Occurence
### * -> Zero or more Occurence 

```python
re.sub('^<td>.*">','',arr[0])
```

```python
arr[0]
```

```python
re.sub('^<td>.*">|</a></i></td>|</b>|</a>|</a></b></i></td>|</i></td>|</td>','',arr[1])
```

```python
for i in arr:
  print(re.sub('^<td>.*">|</a></i></td>|</b>|</a>|</a></b></i></td>|</i></td>|</td>|<td>|</span></span></span>', '', str(i)))
```
